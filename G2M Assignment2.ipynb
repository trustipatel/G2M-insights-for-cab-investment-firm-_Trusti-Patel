## 3. Merge all the datasets

## 4. Correlation Analysis

```python
dataset_corr = f_dataset.corr()
dataset_corr
```

## 5. Heatmap to see the correlation

```python
plt.figure(figsize = (16, 9))

# take upper correlation matrix
matrix = np.triu(dataset_corr)

# Generate heatmap correlation
ax = sb.heatmap(dataset_corr, mask = matrix, cmap = 'rainbow', annot = True, linewidth = 1.5)
# Set the title etc
plt.title('Correlation Heatmap of "G2M Insight for Cab Investment", fontsize = 20')

# Set the size of text
sb.set(font_scale = 1.2)
```

#### Observation :  Based on heatmap results we can see that the (Km Travelled & Cost of Trip, Population & Users) have the highest correlation.

## 13. Profit Margin of Cab companies 

```python
company = f_dataset.groupby('Company')
price_charged = company['Price Charged'].mean()
cost_of_trip = company['Cost of Trip'].mean()
ct = cost_of_trip.index
ctv = cost_of_trip.values
pc = price_charged.values

plt.figure(figsize = (12, 6))
plt.bar(ct, pc, label="Revenue")
plt.bar(ct, ctv,label="Profit")
plt.title('Profit Margin')
plt.ylabel('Price Charged')
plt.xlabel('Cost of Trip')
plt.legend()
plt.show()
```

#### Observation : As we see Yellow Cab has a higher Profit Margin than Pink Cab.
